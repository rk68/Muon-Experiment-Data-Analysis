```python
# Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
# Read csv file
data_df = pd.read_csv('labs calibration 3.csv', names=["input","freq","output"])

# Drop first row
data_df = data_df.drop(0, axis=0)
data_df.head()
```

```python
# Assign columns to variables
input_voltage = pd.to_numeric(data_df["input"])
frequency = pd.to_numeric(data_df["freq"])
output_voltage = pd.to_numeric(data_df["output"])

# Calculate the gain
gain = output_voltage/input_voltage 

```

```python
# Calculate time from frequency
time = 1/(frequency*1000)
```

```python

plt.figure(figsize=(10,10))
plt.title('Amplifier Frequency Response', fontsize=20)
plt.plot(frequency, (gain), 'o')
plt.xlabel('Frequency (kHz)', fontsize=20)
plt.ylabel('Gain', fontsize=20)
plt.grid()
```

```python

plt.figure(figsize=(10,10))
plt.title('Amplifier Frequency Response', fontsize=20)
plt.plot(frequency, (gain*0.1/21), 'o')
plt.xlabel('Frequency (kHz)', fontsize=20)
plt.ylabel('Amplifier output voltage (V)', fontsize=20)
plt.grid()
```

```python
# Plot gain against frequency

plt.figure(figsize=(10,10))
plt.title('Gain against time', fontsize=20)
plt.plot(time/1000, gain, 'o')
plt.xlabel('time')
plt.ylabel('Gain (Vout/Vin)')
plt.grid()
```

```python
# Plot with fit
plt.figure(figsize=(10,10))
plt.title('Gain against time', fontsize=20)
#plt.plot(time, monoExp(frequency, m, t, b), '--', label="fitted")
plt.plot(time, gain, '.')
#plt.legend()
plt.ylabel('Gain, Vout/Vin')
plt.xlabel('Time, s')
plt.grid()
```

```python
plt.figure(figsize=(10,10))
plt.title('Gain against time (range: 0s to 0.004s)', fontsize=20)
#plt.plot(time, monoExp(frequency, m, t, b), '--', label="fitted")
plt.plot(time, gain, '.')
plt.xlim(-0.0001,0.0004)
#plt.legend()
plt.ylabel('Gain, Vout/Vin', fontsize=20)
plt.xlabel('Time, s', fontsize=20)
plt.grid()
```

```python
plt.figure(figsize=(10,7))
plt.title('Gain against time', fontsize=20)
#plt.plot(time, monoExp(frequency, m, t, b), '--', label="fitted")
plt.plot(time/(1e-6), gain*21, '.', label = "Gain Data", color='blue')
#plt.legend()
plt.ylabel('Gain', fontsize=20)
plt.xlabel('Time (µs)', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.axvline(x=0.0000523/(1e-6), color = 'red')
plt.axvline(x=0.00001/(1e-6), color = 'red', label = 'Bandwidth Limits')
plt.xscale("log")
plt.legend(fontsize=15)
plt.grid()
```

```python
plt.figure(figsize=(10,10))
plt.title('Gain against time)', fontsize=20)
#plt.plot(time, monoExp(frequency, m, t, b), '--', label="fitted")
plt.plot(time/(1e-6), gain, '.', label = "data point")
#plt.legend()
plt.ylabel('Gain', fontsize=20)
plt.xlabel('Time (µs)', fontsize=20)
plt.axvline(x=0.0000523/(1e-6), color = 'red')
plt.axvline(x=0.00001/(1e-6), color = 'red', label = 'Bandwidth')
plt.xlim([0,10])
plt.legend()
plt.grid()
```
